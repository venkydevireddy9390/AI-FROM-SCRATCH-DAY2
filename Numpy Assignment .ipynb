<h4> <b> Statistical Operations </b> </h4>

```python
# Creating a New Array
array_qwe=np.arange(0,12)
array_qwe
xyz=array_qwe.reshape(3,4)
xyz
```

```python
# Sum Operations
# Produce total sum, row wise sum (first elements), columns wise sum
total_sum=xyz.sum()
print(total_sum)
rows_sum=xyz.sum(axis=1)
print(rows_sum)
col_sum=xyz.sum(axis=0)
print(col_sum)
```
